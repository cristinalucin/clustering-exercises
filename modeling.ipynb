```python
new_col_names = []
for col in df.columns:
    new_col_names.append(col.lower().replace('.','_'))

new_col_names
```

```python
#rename columns
df.columns = new_col_names
```

```python
df.head()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, hue='species')
```

```python
#Making new Df with these columns
X= df[['sepal_length', 'sepal_width']]
```

```python
X
```

### Clustering on features, NOT with target variable

```python
#Modeling
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
clusters= kmeans.predict(X)
```

```python
clusters
```

```python
df['cluster'] = clusters
df.head()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='cluster', data=df, palette='tab10')
```

```python
kmeans.cluster_centers_
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)
centroids
```

```python
plt.figure(figsize=(14, 9))

for cluster, subset in df.groupby('cluster'):
    plt.scatter(subset.sepal_width, subset.sepal_length, label='cluster ' + str(cluster), alpha=.6)

centroids.plot.scatter(y='sepal_length', x='sepal_width', c='black', marker='x', s=1000, ax=plt.gca(), label='centroid')

plt.legend()
plt.xlabel('sepal_width')
plt.ylabel('sepal_length')
plt.title('Visualizing Cluster Centers')
```

```python
#Elbow Method
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(9, 6))
    pd.Series({k: KMeans(k).fit(X).inertia_ for k in range(2, 12)}).plot(marker='x')
    plt.xticks(range(2, 12))
    plt.xlabel('k')
    plt.ylabel('inertia')
    plt.title('Change in inertia as k increases')
```

### Looks like the "elbow" is between 3 and 5, I would test in this range

```python
fig, axs = plt.subplots(2, 2, figsize=(13, 13), sharex=True, sharey=True)

for ax, k in zip(axs.ravel(), range(2, 6)):
    clusters = KMeans(k).fit(X).predict(X)
    ax.scatter(X.sepal_length, X.sepal_width, c=clusters)
    ax.set(title='k = {}'.format(k), xlabel='sepal length', ylabel='sepal width')
```

### 2. Use the techniques discussed in this lesson, as well as the insights gained from the exploration exercise to perform clustering on the mall customers dataset. Be sure to visualize your results!

```python
import wrangle_mall as wm
```

```python
df = wm.acquire_mall_customers()
```

```python
df.head()
```

```python
#Split the data
train, test, split = wm.train_validate_test_split(df)
```

```python
train.head()
```

```python
#make dummy variables from gender
train = pd.get_dummies(train, columns=['gender'], drop_first=True)
```

```python
train.head()
```

```python
sns.pairplot(train)
```

```python
X = train[['spending_score', 'annual_income']]
```

```python
X
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

clusters = kmeans.predict(X)
clusters
```

```python
train['cluster'] = clusters
train.head()
```

```python
sns.scatterplot(x='spending_score', y='annual_income', hue='cluster', data=train, palette='tab10')
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)
centroids
```

```python
plt.figure(figsize=(14, 9))

for cluster, subset in train.groupby('cluster'):
    plt.scatter(subset.spending_score, subset.annual_income, label='cluster ' + str(cluster), alpha=.6)

centroids.plot.scatter(y='annual_income', x='spending_score', c='black', marker='x', s=1000, ax=plt.gca(), label='centroid')

plt.legend()
plt.xlabel('spending_score')
plt.ylabel('annual_income')
plt.title('Visualizing Cluster Centers')
```

```python
#Elbow Method
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(9, 6))
    pd.Series({k: KMeans(k).fit(X).inertia_ for k in range(2, 12)}).plot(marker='x')
    plt.xticks(range(2, 12))
    plt.xlabel('k')
    plt.ylabel('inertia')
    plt.title('Change in inertia as k increases')
```

```python
#retrying with four or five clusters
fig, axs = plt.subplots(2, 2, figsize=(13, 13), sharex=True, sharey=True)

for ax, k in zip(axs.ravel(), range(2, 6)):
    clusters = KMeans(k).fit(X).predict(X)
    ax.scatter(X.annual_income, X.spending_score, c=clusters)
    ax.set(title='k = {}'.format(k), xlabel='spending_score', ylabel='annual_income')
```

### Looks like K=5 makes the most logical clustering
